# Pandas In-Built Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

df1 = pd.read_csv('data/df1', index_col=0)
df2 = pd.read_csv('data/df2')
df3 = pd.read_csv('data/df3')
```

```python
df1.head()
# TimeSeries
```

```python
df2.head()
```

```python
df3.head()
```

## Histogram

There are several options.

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df1['A'].plot.hist(bins=30)
```

```python
df3['a'].hist()
```

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=30, alpha=0.5)
```

## Area

```python
df2.plot.area(alpha=0.4)
```

```python
# f = plt.figure()  # is not using jupyter
df3[0:30].plot.area(cmap='rainbow', alpha=0.4, legend=False)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
# plt.show()  # if not using jupyter
```

## Barplot

```python
df2.plot.bar(stacked=True)
```

## Lineplot

```python
df1.plot.line(y='B', figsize=(12, 3), lw=1)
```

```python
df1.plot.line(y='A', figsize=(12, 3), lw=1, color='indianred')
```

## Scatterplot

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*30)
```

```python
df3.plot.scatter(x='a', y='b', s=50, c='red', figsize=(12, 3), marker='o', xlim=[-0.2, 1.2], ylim=[-0.2, 1.2])
plt.grid(True)
```

## Boxplot

```python
df2.plot.box()
```

```python
df3[['a', 'b']].plot.box(color='red')
```

## Hexplot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25 ,cmap='rainbow')
```
